```python
import pandas as pd
import numpy as np

# Define the zones and areas
zones = ["North Zone", "Central Zone", "West Zone", "South Zone", "Southwest Zone", "Southeast Zone", "Northwest Zone"]
areas = [
    # List of areas for each zone
    ["Tondiarpet", "Royapuram", "Thiru-Vi-Ka Nagar", "Manali", "Washermanpet", "Choolai", "Madhavaram", "Sembium", "Retteri"],
    ["George Town", "Sowcarpet", "Triplicane", "Mylapore", "Egmore", "Nungambakkam", "Purasawalkam", "Chintadripet", "Vepery"],
    ["Anna Nagar", "Kilpauk", "Mogappair", "Ambattur", "Virugambakkam", "Koyambedu", "Valasaravakkam", "Avadi", "Ayyapakkam", "Maduravoyal", "Nolambur"],
    ["Adyar", "Besant Nagar", "Thiruvanmiyur", "Velachery", "Guindy", "Kotturpuram", "Madipakkam", "Pallikaranai", "Perungudi", "Sholinganallur", "Karapakkam", "Neelankarai"],
    ["Saidapet", "T. Nagar", "Kodambakkam", "Teynampet", "West Mambalam", "Ashok Nagar", "Alwarpet", "Nandanam"],
    ["Sholinganallur", "Perungudi", "Thoraipakkam", "Neelankarai", "Palavakkam", "Kottivakkam", "Injambakkam", "Tambaram", "Velachery", "Medavakkam", "Perumbakkam"],
    ["Madhavaram", "Perambur", "Kolathur", "Puzhal", "Sembium", "Retteri", "Villivakkam", "Ayanavaram", "Korattur"]
]

# Define the features
features = ["Water Flow Rate", "Pressure", "Temperature", "Pipe Diameter", "Distance from Pump Station"]

# Simulate data with leaks
data = []
for zone, area_list in zip(zones, areas):
    for area in area_list:
        for _ in range(np.random.randint(50, 100)):  # Simulate multiple samples for each area
            leak = np.random.choice([0, 1], p=[0.9, 0.1])  # Simulate leak occurrence with 10% probability
            features_data = [np.random.normal(loc=100, scale=10) for _ in range(len(features))]  # Simulate feature data
            data.append([zone, area, leak] + features_data)

# Create a DataFrame
df = pd.DataFrame(data, columns=["Zone", "Area", "Leak"] + features)

# Save simulated data to a CSV file
df.to_csv("simulated_data_with_leaks.csv", index=False)
```
