```python
import pandas as pd
import numpy as np

# Load sample data from CSV
df = pd.read_csv("D:/Water Management System/data/water consumption data/overall_data.csv")

# Define function to generate simulated data with some values above upper bounds for different areas
def generate_simulated_data(row):
    # Generate random water consumption within upper bounds
    water_consumption = {
        'Schools': np.random.randint(0, row['Schools Upper Bound (liters/day)']),
        'Colleges': np.random.randint(0, row['Colleges Upper Bound (liters/day)']),
        'Canteens': np.random.randint(0, row['Canteens Upper Bound (liters/day)']),
        'Hotels': np.random.randint(0, row['Hotels Upper Bound (liters/day)']),
        'Malls': np.random.randint(0, row['Malls Upper Bound (liters/day)']),
        'Hospitals': np.random.randint(0, row['Hospitals Upper Bound (liters/day)']),
        'Railway Stations': np.random.randint(0, row['Railway Stations Upper Bound (liters/day)']),
        'Offices': np.random.randint(0, row['Offices Upper Bound (liters/day)']),
        'Stadiums': np.random.randint(0, row['Stadiums Upper Bound (liters/day)']),
        'Theaters': np.random.randint(0, row['Theaters Upper Bound (liters/day)']),
        'Bus Stations': np.random.randint(0, row['Bus Stations Upper Bound (liters/day)']),
        '1 Bedroom House': np.random.randint(0, row['1 Bedroom Water Consumption (Upper)']),
        '2 Bedroom House': np.random.randint(0, row['2 Bedroom Water Consumption (Upper)']),
        '3 Bedroom House': np.random.randint(0, row['3 Bedroom Water Consumption (Upper)']),
        '4 Bedroom House': np.random.randint(0, row['4 Bedroom Water Consumption (Upper)']),
        'Industries': np.random.randint(0, row['Total Daily Water Consumption (Upper)']),
    }
    
    # Adjust values for specific areas to simulate data above upper bounds
    if row['Area Name'] == 'Royapuram':
        water_consumption['Industries'] += 10453 
    if row['Area Name'] == 'Sowcarpet':
        water_consumption['Stadiums'] += 5231 
    if row['Area Name'] == 'Egmore':
        water_consumption['Theaters'] += 16060  
    if row['Area Name'] == 'Guindy':
        water_consumption['Colleges'] += 20400  
    if row['Area Name'] == 'Saidapet':
        water_consumption['Malls'] += 10436  
    if row['Area Name'] == 'Tambaram':
        water_consumption['Bus Stations'] += 5638
    if row['Area Name'] == 'Perambur':
        water_consumption['2 Bedroom House'] += 43260

    return pd.Series(water_consumption)

# Apply function to generate simulated data for each row
simulated_data = df.apply(generate_simulated_data, axis=1)

# Concatenate 'Zone Name' and 'Area Name' with simulated data
result = pd.concat([df[['Zone Name', 'Area Name']], simulated_data], axis=1)

# Save simulated data to CSV
result.to_csv('real_time_water_consumption_data.csv', index=False)

# Display the simulated data
print(result.head())
```
