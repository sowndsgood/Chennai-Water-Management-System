```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the water usage data from the CSV file
csv_filename = 'water_usage_data.csv'
usage_df = pd.read_csv(csv_filename)

# Convert the 'Timestamp' column to datetime format
usage_df['Timestamp'] = pd.to_datetime(usage_df['Timestamp'])

# Calculate total water usage for each device
total_usage = usage_df.groupby('Device Name')['Usage (liters)'].sum()

# Plot pie chart for proportion of total water usage for each device
plt.figure(figsize=(8, 8))
total_usage.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
plt.axis('equal')
plt.title('Proportion of Total Water Usage for Each Device')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the water usage data from the CSV file
csv_filename = 'water_usage_data.csv'
usage_df = pd.read_csv(csv_filename)

# Convert the 'Timestamp' column to datetime format
usage_df['Timestamp'] = pd.to_datetime(usage_df['Timestamp'])

# Plot scatter plot with points joined by lines for each device separately
for device_name, device_data in usage_df.groupby('Device Name'):
    plt.figure(figsize=(10, 6))
    plt.plot(device_data['Timestamp'], device_data['Usage (liters)'], marker='o', linestyle='-')
    plt.xlabel('Timestamp')
    plt.ylabel('Usage (liters)')
    plt.title(f'Water Usage Over Time for Device: {device_name}')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
