```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the water usage data from the CSV file
csv_filename = 'april_water_usage_data.csv'
usage_df = pd.read_csv(csv_filename)

# Convert the 'Timestamp' column to datetime format
usage_df['Timestamp'] = pd.to_datetime(usage_df['Timestamp'])

# Display the first few rows of the data
usage_df.head()

# Summary statistics
summary_stats = usage_df.groupby('Device Name')['Usage (liters)'].describe()

# Display the summary statistics
summary_stats

# Plot bar chart for total water usage of each device
plt.figure(figsize=(10, 6))
usage_df.groupby('Device Name')['Usage (liters)'].sum().plot(kind='bar', color='skyblue')
plt.xlabel('Device Name')
plt.ylabel('Total Usage (liters)')
plt.title('Total Water Usage for Each Device in April')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot pie chart for proportion of total water usage for each device
plt.figure(figsize=(8, 8))
usage_df.groupby('Device Name')['Usage (liters)'].sum().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.tab10.colors)
plt.axis('equal')
plt.title('Proportion of Total Water Usage for Each Device in April')
plt.tight_layout()
plt.show()

# Plot scatter plot with points joined by lines for each device separately
for device_name, device_data in usage_df.groupby('Device Name'):
    plt.figure(figsize=(10, 6))
    plt.plot(device_data['Timestamp'], device_data['Usage (liters)'], marker='o', linestyle='-')
    plt.xlabel('Timestamp')
    plt.ylabel('Usage (liters)')
    plt.title(f'Water Usage Over Time for Device: {device_name} (April)')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Write summary statistics to a text file
with open('usage_summary.txt', 'w') as f:
    f.write(summary_stats.to_string())

print("Summary statistics saved to 'usage_summary.txt'.")
```
